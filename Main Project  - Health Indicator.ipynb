<h3> Visual 7 - State-Wise Multivariate Correlation of Abortion Metrics and Maternal Mortality - Bubble Plot</h3>

```python
import matplotlib.pyplot as plt

abortion = df.groupby('State Name (state_name)').agg({
    'Abortion Spontaneous (abortion_spontaneous)': 'sum',
    'Mtp Up to 12 Weeks of Pregnancy (mtp_up_to_12_weeks_of_pregnancy)': 'sum',
    'Mtp More Than 12 Weeks of Pregnancy (mtp_more_than_12_weeks_of_pregnancy)': 'sum',
    'Number of Maternal Deaths due to Abortion (number_of_maternal_deaths_due_to_abortion)': 'sum'
})

plt.figure(figsize=(12,8))

scatter = plt.scatter(
    abortion['Abortion Spontaneous (abortion_spontaneous)'],
    abortion['Mtp Up to 12 Weeks of Pregnancy (mtp_up_to_12_weeks_of_pregnancy)'],
    s=abortion['Mtp More Than 12 Weeks of Pregnancy (mtp_more_than_12_weeks_of_pregnancy)']*5,
    c=abortion['Number of Maternal Deaths due to Abortion (number_of_maternal_deaths_due_to_abortion)'],
    cmap='Reds',
    alpha=0.7
)

plt.colorbar(scatter, label='Maternal Deaths Due to Abortion')

plt.xlabel('Spontaneous Abortions')
plt.ylabel('MTP Up to 12 Weeks')
plt.title('State-wise Abortion and Maternal Health Analysis')

for state in abortion.index:
    plt.text(
        abortion.loc[state, 'Abortion Spontaneous (abortion_spontaneous)'],
        abortion.loc[state, 'Mtp Up to 12 Weeks of Pregnancy (mtp_up_to_12_weeks_of_pregnancy)'],
        state,
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

<h3>Key Insights</h3>
<ul>
    <li><strong>Madhya Pradesh Outlier in MTPs:</strong> Madhya Pradesh records the highest number of Medical Terminations of Pregnancy (MTP) up to 12 weeks, nearing 14,000 cases, despite having lower spontaneous abortion rates (~30,000).</li>
    <li><strong>Tamil Nadu Leads in Spontaneous Abortions:</strong> Tamil Nadu displays the highest volume of spontaneous abortions, exceeding 330,000 cases, while maintaining a moderate MTP count around 4,000.</li>
    <li><strong>Uttar Pradesh High Volume Cluster:</strong> Uttar Pradesh shows a unique footprint with high spontaneous abortions (~125,000) but a relatively low number of reported early MTPs.</li>
    <li><strong>Low Maternal Mortality Across High Volumes:</strong> The light shading across almost all visible bubbles suggests reported maternal deaths due to abortion remain at the lowest end of the scale (0 to 1) for these specific states.</li>
    <li><strong>Dense Lower-Left Concentration:</strong> The majority of states and Union Territories cluster heavily near the origin point, indicating low absolute reporting for both metrics.</li>
</ul>

<h3>Visual 10 — Comparative Evaluation of BCG, DPT1, and OPV1 Performance by State - Scatter Plot</h3>

```python
import matplotlib.pyplot as plt

state = df.groupby('State Name (state_name)').agg({
    'Child Immunisation Bcg (child_immunisation_bcg)':'sum',
    'Child Immunisation Dpt1 (child_immunisation_dpt1)':'sum',
    'Child Immunisation Measles1St Dose (child_immunisation_measles1st_dose)':'sum',
    'Child Immunisation Opv1 (child_immunisation_opv1)':'sum'
})

plt.figure(figsize=(12,8))

scatter = plt.scatter(
    state['Child Immunisation Bcg (child_immunisation_bcg)'],
    state['Child Immunisation Dpt1 (child_immunisation_dpt1)'],
    s=state['Child Immunisation Measles1St Dose (child_immunisation_measles1st_dose)']/50,
    c=state['Child Immunisation Opv1 (child_immunisation_opv1)'],
    cmap='plasma',
    alpha=0.8
)

plt.colorbar(scatter, label='OPV1 Vaccinations')

for i, txt in enumerate(state.index):
    plt.text(
        state.iloc[i, 0],
        state.iloc[i, 1],
        txt,
        fontsize=8
    )

plt.xlabel('BCG Vaccinations')
plt.ylabel('DPT1 Vaccinations')
plt.title('State-wise Multivariate Immunisation Analysis')

plt.tight_layout()
plt.show()
```
